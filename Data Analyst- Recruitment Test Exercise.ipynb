# Task B

Lok_Sabha- http://164.100.47.194/Loksabha/Members/MemberSearch.aspx
List of all the members of the Lok sabha is to be extracted.
(Plus point if you extract the links of their individual pages as well)

```python
from email import header
import requests
from bs4 import BeautifulSoup
import pandas as pd 
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
url='http://loksabhaph.nic.in/Members/AlphabeticalList.aspx'
r=requests.get(url,headers=headers)
soup=BeautifulSoup(r.content,'html.parser')
#print(soup.prettify())
```

```python
trs=soup.find_all('tr',class_="odd")
all_data=[]
links=[]
for tds in trs:
    for td in tds:
         rows=(td.text)
         all_data.append(rows)
    for a in tds.find_all('a', href=True):
        b="http://loksabhaph.nic.in/Members/"+a['href']
        links.append(b)
#print(links)
no_dup_link=[]
for i in links:
    if i not in no_dup_link:
        #m="http://loksabhaph.nic.in/"+i
        no_dup_link.append(i)
details = []
detail=[]
for data in all_data:
    details.append(data.replace("\n", "").replace("\r","").replace('  ',"").replace('xa0',""))
for x in details:
    if x=='':
        pass
    else:
        detail.append(x)
#print(detail[0::4])
data=pd.DataFrame()
no_=detail[::4]
data['No.']=detail[::4]
data['Names']=detail[1::4]
data['Party Name']=detail[2::4]
data['	Constituency (State)']=detail[3::4]
data['profile Links']=no_dup_link
data
#data.to_csv("List of all the members.csv")

```

# Task C

MCA- https://www.mca.gov.in/content/mca/global/en/contact-us/official-
liquidators.html

List of all the official liquidators as well as all the information about them should be
extracted.
(Plus point if you extract the Geo-Location links as well)

```python
from email import header
import requests
from bs4 import BeautifulSoup
import pandas as pd 
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
url='https://www.mca.gov.in/content/mca/global/en/contact-us/official-liquidators.html'
r=requests.get(url,headers=headers)
soup=BeautifulSoup(r.content,'html.parser')
#print(soup.prettify())
```

```python
trs=soup.find_all('tr',class_="view table-values dark-theme diff-designation")
all_data=[]
links=[]
for tds in trs:
    #('a', href=True)
    #name=td.text
    #names.append(name)
    b=tds.a['href']
    links.append(b)
    for td in tds:
        row=td.text
        all_data.append(row)
        #print(row)
#all_data
details = []
detail=[]
for data in all_data:
    details.append(data.replace("\n", "").replace("\r","").replace('  ',"").replace('xa0',""))
for x in details:
    if x=='':
        pass
    else:
        detail.append(x)
#print(detail)[1:-1:4])
name= detail[0:-1:4]
#print(name)
desig=detail[1:-1:4]
contact=detail[2:-1:4]
address=detail[3::4]
add=[]
for x in address:
    add.append(x.replace('View on Map',''))
#print(add)
data_new=pd.DataFrame()
data_new['Name']=name
data_new['Designation']=desig
data_new['Contact']=contact
data_new['Address']=add
data_new['Geo0-Location Link']=links
data_new
#data_new.to_csv("mcu.csv")
```
